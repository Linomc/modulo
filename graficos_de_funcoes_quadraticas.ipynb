### Achei interessante a pergunta no grupo de **GEOMETRIA RAIZ**. Vou apresentar computacionalmente, a justificativa gráfica do resultado da equação por meio de função, acho que fica mais intuitivo.

### Usando o exemplo da equação do colega que respondeu a pergunta de outro colega, o exemplo usado foi esse: ->>> **√x²=|x|**

```python
from sympy.core.symbol import Symbol
x = Symbol('x') # aqui foi definida a variável indepente, que é também nosso domínio, que chamamos de x
```

```python
y = (x**2)**(1/2) # função
```

```python
from sympy.plotting.plot import plot
plot(y, title = 'gráfico', xlabel = 'Domínio', ylabel = 'Imagem', xlim = [-3,3], ylim = [-3,3])
```
